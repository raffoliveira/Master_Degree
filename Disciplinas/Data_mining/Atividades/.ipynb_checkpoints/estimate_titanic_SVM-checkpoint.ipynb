## Importando pacotes

## Carregando os dados

## Pré-processamento

### Remoção de atributos

Alguns atributos serão removidos devido sua alta cardinalidade, o que pode apresentar dificuldade ao modelo em identificar tanto padrões, impossibilitando a generalização do modelo

```python
[f'Attribute {i} has {len(train[i].unique())} unique values.' for i in train.columns]
```

Como podemos observar os atributos com maiores cardinalidade são:
+ PassengerId
+ Name
+ Ticket
+ Cabin

Por isso, esses atributos serão removidos

```python
train_r = train.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
test_r = test.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
```

### Tratamento de valores ausentes

```python
#verificando a quantidade de valores ausentes
train_r.isnull().sum()
```

```python
#verificando a quantidade de valores ausentes
test_r.isnull().sum()
```

```python
#verificando o percentual de valores ausentes
train_r.isna().sum()/len(train_r)
```

```python
#verificando o percentual de valores ausentes
test_r.isna().sum()/len(test_r)
```

Observamos que o atributo **Age** contém 177 e 86 de valores ausentes no conjunto de treino e teste, respectivamente. O atributo **Embarked** tem 2 valores ausentes no conjunto de treino e 1 valor ausente para o atributo **Fare** no conjunto de teste.

Para tratamento, utilizarei a técnica de imputação **Multiple imputation by chained equations (MICE)** para o atributo **Age**. Para os atributos **Embarked** e **Fare** será utilizado a remoção dos valores ausentes, uma vez que a presença é bem pequena.

A técnica MICE, em vez de utilizar apenas uma imputação, preenche os valores ausentes utilizando várias imputações de forma iterativa até uma conversão. 

Mais informações sobre essa técnica pode ser encontrada neste [artigo](https://www.jstatsoft.org/article/view/v045i03/v45i03.pdf).

Abaixo segue uma animação do funcionamento da técnica para três variáveis. Referência: [medium](https://medium.com/@ofirdi/mice-is-nice-but-why-should-you-care-e66698f245a3)

<img src="mice.gif" width=600 height=600 />

```python
#removendo valores ausentes nos atributos Embarked e Fare
train_r.dropna(subset=['Embarked'], inplace=True)
test_r.dropna(subset=['Fare'], inplace=True)
```

```python
train_r.isnull().sum()
```

```python
test_r.isnull().sum()
```

O MICE não aceita atributos categóricos. Por isso, realizarei o one hot encoding nos atributos **Sex** e **Embarked** primeiro para posteriormente aplicar o MICE

### Conversão de atributos categóricos

Conversão utilizando a One Hot Enconding para os atributos **Sex** e **Embarked**

```python
#aplicando o one hot enconding
train_r = pd.get_dummies(train_r, columns=['Sex','Embarked'])
test_r = pd.get_dummies(test_r, columns=['Sex','Embarked'])
```

### Aplicação da técnica MICE

```python
#aplicando o MICE para o atributo Age
imputed_values_train = mice(train_r.values)
imputed_values_test = mice(test_r.values)
```

```python
#convertendo o resultado em dataframe
train_r = pd.DataFrame(imputed_values_train, columns=train_r.columns)
test_r = pd.DataFrame(imputed_values_test, columns=test_r.columns)
```

```python
#verificando se há valores ausentes
train_r.isnull().sum()
```

```python
test_r.isnull().sum()
```

### Remoção de duplicatas

```python
#removendo duplicatas
train_r.drop_duplicates(inplace=True)
test_r.drop_duplicates(inplace=True)
```

## Construindo o modelo

```python
#dividindo o conjunto de treino em conjunto de atributos e classe (alvo)
train_X = train_r.drop(['Survived'],axis=1)
train_y = train_r['Survived']
```

```python
#construção do gridSearch
def gridSearchCV(X,y,kind_grid):    

    #especificando os hiper-parametros de busca 
    #o parametro C é configurado com uma lista randômica uniforme de 50 valores
    #parameters = {'SVM__C':sp.stats.uniform(scale=0.5), 'SVM__kernel':('poly','rbf'), 'SVM__gamma':['scale', 'auto',0.1,1,10]}
    parameters = {'SVM__C':np.random.rand(1,50)[0], 'SVM__kernel':['rbf'], 'SVM__gamma':['scale','auto',0.1,1]}
    
    #pipeline para realizar normalização dos atributos
    norm_transformer = Pipeline(steps=[('minMaxScaler', MinMaxScaler())])

    #seleção dos atributos que serão normalizados
    norm_feat = ["Pclass", "Age", "SibSp", "Parch", "Fare"]

    #compondo o pipeline de aplicação da normalização
    preprocessor = ColumnTransformer(transformers=[
        ('norm_feat', norm_transformer, norm_feat)
    ])

    #criando o modelo atraves de pipeline com o preprocessamento e aplicação do classificador
    model = Pipeline(steps=[('processor', preprocessor),   
                            ('SVM', SVC())])

    #criando um dicionario com as métricas que serão avaliadas
    scores_ = {'accuracy' : metrics.make_scorer(metrics.accuracy_score),
               'recall'   : metrics.make_scorer(metrics.recall_score, average='weighted', zero_division=0),
               'precision': metrics.make_scorer(metrics.precision_score, average='weighted', zero_division=0),
               'f1'       : metrics.make_scorer(metrics.fbeta_score, beta = 2, average='weighted' )} # beta>1 favors recall
    
    #configurando o 10-fold
    #este modelo de Kfold repete o StratifiedKfold (preserva a distribuição das classes durante os splits) n vezes com diferente 
    #randomização em cada repetição
    fold = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=1)

    #instanciamento do GridSearch com o modelo criado e o conjunto de parametros
    #parametro *scoring* calcula todas as metricas definidas no dicionario
    #parametro *refit* faz o gridsearch escolher o melhor parametro de acordo com essa metrica, ou seja, ele não consegue 
    #decidir dentre todas as metricas calculadas. É preciso escolher apenas uma metrica
    #paramentro *cv* usa a estrategia de split usando fold. Quando setamos um numero inteiro, o grid utiliza o StratifiedKFold, o qual
    # faz o split de modo que os folds preservem o percentual de distribuição das classes alvos
    if kind_grid == 'gridSearch':        
        grid = GridSearchCV(model, 
                            param_grid=parameters, 
                            cv=fold, 
                            n_jobs=-1, 
                            scoring=scores_, 
                            refit='accuracy',
                            return_train_score=True)
    else:        
        grid = RandomizedSearchCV(model, 
                                  param_distributions=parameters, 
                                  n_iter=100, 
                                  cv=fold, 
                                  n_jobs=-1,
                                  scoring=scores_, 
                                  refit='accuracy',
                                  return_train_score=True)

    #treinametro do modelo com o conjunto de teste
    grid.fit(X,y)
#     grid_randomized.fit(X, y)
    
    #imprime o melhor score 
    print(f'Best_score: {grid.best_score_}')
#     print(f'Best_score: {grid_randomized.best_score_}')
    
    #imprime o melhor conjunto de parametro
    print(f'Best_param: {grid.best_params_}')
#     print(f'Best_param: {grid_randomized.best_params_}')
    
    #realiza um resumo dos resultados das iterações
    result = pd.DataFrame(grid.cv_results_)[['params',
                                             'mean_train_accuracy',
                                             'std_train_accuracy',
                                             'mean_test_accuracy',
                                             'std_test_accuracy']].sort_values(by=['mean_test_accuracy'], ascending=False)  
    return result, grid
```

## Testando o modelo com GridSearchCV e RandomizedSearchCV

Faço o teste para verificar se há diferença entre as técnicas de busca de hiper parametros

```python
#testando GridSearchCV
result,grid = gridSearchCV(train_X,train_y,'gridSearch')
```

```python
#imprimindo os resultados do GridSearchCV. Há 200 linhas pois dependente do numero de combinações de parametros (50*1*4=200)
result
```

```python
#visualizando a matriz de confusão do conjunto de treino
train_pred = grid.predict(train_X)
sns.heatmap(confusion_matrix(train_y,train_pred), annot=True, fmt='.0f')
plt.xlabel('Pred_label')
plt.ylabel('True_label')
```

Observando a matriz de confusão, percebe-se que o modelo classificou erroneamente **14,4%** para a classe **0** e **50,62%** para a classe **1**. Logo, vemos que o modelo erra mais da metade para a classe **1**, o que interfere na acurácia do modelo.

```python
#testando RandomizedSearchCV
result_rand, grid_rand = gridSearchCV(train_X,train_y,'randomizedSearch')
```

```python
#imprimindo os resultados RandomizedSearchCV. Há 100 linhas pois o modelo realiza 100 iterações
result_rand
```

```python
#visualizando a matriz de confusão do conjunto de treino
train_pred_rand = grid_rand.predict(train_X)
sns.heatmap(confusion_matrix(train_y,train_pred_rand), annot=True, fmt='.0f')
plt.xlabel('Pred_label')
plt.ylabel('True_label')
```

A matriz de confusão utilizando a técnica RandomizedSearchCV não muda significamente em comparação com a GridSearchCV obtendo praticamente os mesmos valores.

Os resultados entre os dois modelos de searchCV foram basicamente semelhantes, mostrando acurácia de 71% para o conjunto de treino e 70% para o conjunto de teste. Lembrando, que o conjunto inicial de treino foi dividido em treino e teste para a criação do modelo de classificação. Mesmo o GridSearch realizando a busca em todas as combinações possíveis de parâmetros e o RandomizedSearchCV realizando apenas de forma randômica (de acordo com o número de iterações), não houve mudança considerável nos resultados finais dos dois métodos.
