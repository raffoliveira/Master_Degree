## Importando pacotes

Alguns atributos serÃ£o removidos devido sua alta cardinalidade, o que pode apresentar dificuldade ao modelo em identificar tanto padrÃµes, impossibilitando a generalizaÃ§Ã£o do modelo

Observamos que o atributo **Age** contÃ©m 177 e 86 de valores ausentes no conjunto de treino e teste, respectivamente. O atributo **Embarked** tem 2 valores ausentes no conjunto de treino e 1 valor ausente para o atributo **Fare** no conjunto de teste.

Para tratamento, utilizarei a tÃ©cnica de imputaÃ§Ã£o **Multiple imputation by chained equations (MICE)** para o atributo **Age**. Para os atributos **Embarked** e **Fare** serÃ¡ utilizado a remoÃ§Ã£o dos valores ausentes, uma vez que a presenÃ§a Ã© bem pequena.

A tÃ©cnica MICE, em vez de utilizar apenas uma imputaÃ§Ã£o, preenche os valores ausentes utilizando vÃ¡rias imputaÃ§Ãµes de forma iterativa atÃ© uma conversÃ£o. 

Mais informaÃ§Ãµes sobre essa tÃ©cnica pode ser encontrada neste [artigo](https://www.jstatsoft.org/article/view/v045i03/v45i03.pdf).

Abaixo segue uma animaÃ§Ã£o do funcionamento da tÃ©cnica para trÃªs variÃ¡veis. ReferÃªncia: [medium](https://medium.com/@ofirdi/mice-is-nice-but-why-should-you-care-e66698f245a3)

<img src="mice.gif" width=600 height=600 />

```python
#removendo valores ausentes nos atributos Embarked e Fare
train_r.dropna(subset=['Embarked'], inplace=True)
test_r.dropna(subset=['Fare'], inplace=True)
```

```python
train_r.isnull().sum()
```

```python
test_r.isnull().sum()
```

O MICE nÃ£o aceita atributos categÃ³ricos. Por isso, realizarei o one hot encoding nos atributos **Sex** e **Embarked** primeiro para posteriormente aplicar o MICE

### ConversÃ£o de atributos categÃ³ricos

ConversÃ£o utilizando a One Hot Enconding para os atributos **Sex** e **Embarked**

```python
#aplicando o one hot enconding
train_r = pd.get_dummies(train_r, columns=['Sex','Embarked'])
test_r = pd.get_dummies(test_r, columns=['Sex','Embarked'])
```

### AplicaÃ§Ã£o da tÃ©cnica MICE

```python
#aplicando o MICE para o atributo Age
imputed_values_train = mice(train_r.values)
imputed_values_test = mice(test_r.values)
```

```python
#convertendo o resultado em dataframe
train_r = pd.DataFrame(imputed_values_train, columns=train_r.columns)
test_r = pd.DataFrame(imputed_values_test, columns=test_r.columns)
```

```python
#verificando se hÃ¡ valores ausentes
train_r.isnull().sum()
```

```python
test_r.isnull().sum()
```

### RemoÃ§Ã£o de duplicatas

```python
#removendo duplicatas
train_r.drop_duplicates(inplace=True)
test_r.drop_duplicates(inplace=True)
```

## Construindo o modelo

```python
#dividindo o conjunto de treino em conjunto de atributos e classe (alvo)
train_X = train_r.drop(['Survived'],axis=1)
train_y = train_r['Survived']
```

```python
#construÃ§Ã£o do gridSearch
def gridSearchCV(X,y,kind_grid):    

    #especificando os hiper-parametros de busca 
    #o parametro C Ã© configurado com uma lista randÃ´mica uniforme de 50 valores
    #parameters = {'SVM__C':sp.stats.uniform(scale=0.5), 'SVM__kernel':('poly','rbf'), 'SVM__gamma':['scale', 'auto',0.1,1,10]}
    parameters = {'SVM__C':np.random.rand(1,50)[0], 'SVM__kernel':['rbf'], 'SVM__gamma':['scale','auto',0.1,1]}
    
    #pipeline para realizar normalizaÃ§Ã£o dos atributos
    norm_transformer = Pipeline(steps=[('minMaxScaler', MinMaxScaler())])

    #seleÃ§Ã£o dos atributos que serÃ£o normalizados
    norm_feat = ["Pclass", "Age", "SibSp", "Parch", "Fare"]

    #compondo o pipeline de aplicaÃ§Ã£o da normalizaÃ§Ã£o
    preprocessor = ColumnTransformer(transformers=[
        ('norm_feat', norm_transformer, norm_feat)
    ])

    #criando o modelo atraves de pipeline com o preprocessamento e aplicaÃ§Ã£o do classificador
    model = Pipeline(steps=[('processor', preprocessor),   
                            ('SVM', SVC())])

    #criando um dicionario com as mÃ©tricas que serÃ£o avaliadas
    scores_ = {'accuracy' : metrics.make_scorer(metrics.accuracy_score),
               'recall'   : metrics.make_scorer(metrics.recall_score, average='weighted', zero_division=0),
               'precision': metrics.make_scorer(metrics.precision_score, average='weighted', zero_division=0),
               'f1'       : metrics.make_scorer(metrics.fbeta_score, beta = 2, average='weighted' )} # beta>1 favors recall
    
    #configurando o 10-fold
    #este modelo de Kfold repete o StratifiedKfold (preserva a distribuiÃ§Ã£o das classes durante os splits) n vezes com diferente 
    #randomizaÃ§Ã£o em cada repetiÃ§Ã£o
    fold = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=1)

    #instanciamento do GridSearch com o modelo criado e o conjunto de parametros
    #parametro *scoring* calcula todas as metricas definidas no dicionario
    #parametro *refit* faz o gridsearch escolher o melhor parametro de acordo com essa metrica, ou seja, ele nÃ£o consegue 
    #decidir dentre todas as metricas calculadas. Ã‰ preciso escolher apenas uma metrica
    #paramentro *cv* usa a estrategia de split usando fold. Quando setamos um numero inteiro, o grid utiliza o StratifiedKFold, o qual
    # faz o split de modo que os folds preservem o percentual de distribuiÃ§Ã£o das classes alvos
    if kind_grid == 'gridSearch':        
        grid = GridSearchCV(model, 
                            param_grid=parameters, 
                            cv=fold, 
                            n_jobs=-1, 
                            scoring=scores_, 
                            refit='accuracy',
                            return_train_score=True)
    else:        
        grid = RandomizedSearchCV(model, 
                                  param_distributions=parameters, 
                                  n_iter=100, 
                                  cv=fold, 
                                  n_jobs=-1,
                                  scoring=scores_, 
                                  refit='accuracy',
                                  return_train_score=True)

    #treinametro do modelo com o conjunto de teste
    grid.fit(X,y)
#     grid_randomized.fit(X, y)
    
    #imprime o melhor score 
    print(f'Best_score: {grid.best_score_}')
#     print(f'Best_score: {grid_randomized.best_score_}')
    
    #imprime o melhor conjunto de parametro
    print(f'Best_param: {grid.best_params_}')
#     print(f'Best_param: {grid_randomized.best_params_}')
    
    #realiza um resumo dos resultados das iteraÃ§Ãµes
    result = pd.DataFrame(grid.cv_results_)[['params',
                                             'mean_train_accuracy',
                                             'std_train_accuracy',
                                             'mean_test_accuracy',
                                             'std_test_accuracy']].sort_values(by=['mean_test_accuracy'], ascending=False)  
    return result, grid
```

## Testando o modelo com GridSearchCV e RandomizedSearchCV

FaÃ§o o teste para verificar se hÃ¡ diferenÃ§a entre as tÃ©cnicas de busca de hiper parametros

```python
#testando GridSearchCV
result,grid = gridSearchCV(train_X,train_y,'gridSearch')
```

```python
#imprimindo os resultados do GridSearchCV. HÃ¡ 200 linhas pois dependente do numero de combinaÃ§Ãµes de parametros (50*1*4=200)
result
```

```python
#visualizando a matriz de confusÃ£o do conjunto de treino
train_pred = grid.predict(train_X)
sns.heatmap(confusion_matrix(train_y,train_pred), annot=True, fmt='.0f')
plt.xlabel('Pred_label')
plt.ylabel('True_label')
```

Observando a matriz de confusÃ£o, percebe-se que o modelo classificou erroneamente **14,4%** para a classe **0** e **50,62%** para a classe **1**. Logo, vemos que o modelo erra mais da metade para a classe **1**, o que interfere na acurÃ¡cia do modelo.

```python
#testando RandomizedSearchCV
result_rand, grid_rand = gridSearchCV(train_X,train_y,'randomizedSearch')
```

```python
#imprimindo os resultados RandomizedSearchCV. HÃ¡ 100 linhas pois o modelo realiza 100 iteraÃ§Ãµes
result_rand
```

```python
#visualizando a matriz de confusÃ£o do conjunto de treino
train_pred_rand = grid_rand.predict(train_X)
sns.heatmap(confusion_matrix(train_y,train_pred_rand), annot=True, fmt='.0f')
plt.xlabel('Pred_label')
plt.ylabel('True_label')
```

A matriz de confusÃ£o utilizando a tÃ©cnica RandomizedSearchCV nÃ£o muda significamente em comparaÃ§Ã£o com a GridSearchCV obtendo praticamente os mesmos valores.

Os resultados entre os dois modelos de searchCV foram basicamente semelhantes, mostrando acurÃ¡cia de 71% para o conjunto de treino e 70% para o conjunto de teste. Lembrando, que o conjunto inicial de treino foi dividido em treino e teste para a criaÃ§Ã£o do modelo de classificaÃ§Ã£o. Mesmo o GridSearch realizando a busca em todas as combinaÃ§Ãµes possÃ­veis de parÃ¢metros e o RandomizedSearchCV realizando apenas de forma randÃ´mica (de acordo com o nÃºmero de iteraÃ§Ãµes), nÃ£o houve mudanÃ§a considerÃ¡vel nos resultados finais dos dois mÃ©todos.
