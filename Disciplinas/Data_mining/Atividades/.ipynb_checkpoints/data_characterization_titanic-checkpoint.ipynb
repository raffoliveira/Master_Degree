## e) Correlation between attributes

```python
correlation = df.corr()
```

```python
#The highest correlation = (Survived->Fare) (SibSp->Parch) (Parch->SibSp) (Parch->Fare) (Fare->Survived)
plt.figure(figsize=(9,6))
sns.heatmap(correlation, 
            xticklabels=correlation.columns,
            yticklabels=correlation.columns,
            annot=True)
```
