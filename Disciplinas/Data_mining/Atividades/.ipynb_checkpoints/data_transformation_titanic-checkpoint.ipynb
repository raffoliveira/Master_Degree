## a) Valores ausentes

```python
#mostra as colunas com valores nan
df.isna().sum()
```

observando os atributos que apresentam valores nulos, o atributo **Cabin** apresenta mais de 77% dos valores ausentes. Por isso, esse atributo será excluído

```python
df = df.drop(columns=['Cabin'])
df
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

gerando os valores para os atributos **'Survived', 'Age', 'Fare'** usando a classe InterativeImputer

o estimador padrão é BayesianRidge()

```python
values = df[['Survived', 'Age', 'Fare']].values 
imp = IterativeImputer(initial_strategy='most_frequent', max_iter=20, random_state=0)
values_estimated = np.round(imp.fit_transform(values))
df[['Survived', 'Age', 'Fare']] = values_estimated
```

```python
# encontrando os 2 casos de valores nan para o atributo ['Embarked']
df[df['Embarked'].isna() == True]
```

## b) Suavize os ruídos

Usarei a estratégia de substituir os outliers pelos valores da fronteira externa

```python
# encontrando os valores da barreira externa
age_plot = plt.boxplot(df['Age'])
b = [item.get_ydata() for item in age_plot['whiskers']] # extraindo os limites do boxplot
print(b)
```

```python
# encontrando os valores da barreira externa
sibSp_plot = plt.boxplot(df['SibSp'])
b = [item.get_ydata() for item in sibSp_plot['whiskers']] # extraindo os limites do boxplot
print(b)
```

```python
# encontrando os valores da barreira externa
parch_plot = plt.boxplot(df['Parch'])
b = [item.get_ydata() for item in parch_plot['whiskers']] # extraindo os limites do boxplot
print(b)
```

```python
# encontrando os valores da barreira externa
fare_plot = plt.boxplot(df['Fare'])
b = [item.get_ydata() for item in fare_plot['whiskers']] # extraindo os limites do boxplot
print(b)
```

```python
# substituindo pelos valores da barreira externa
df['Age'] = np.where(df['Age'] < 3, 3, df['Age'])
df['Age'] = np.where(df['Age'] > 54, 54, df['Age'])

df['SibSp'] = np.where(df['SibSp'] > 2, 2, df['SibSp'])

df['Parch'] = np.where(df['Parch'] > 0, 0, df['Parch'])

df['Fare'] = np.where(df['Fare'] > 65, 65, df['Fare'])
```

## d) Transforme atributos categóricos nominais em um vetor numérico

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
X = df[['Pclass','Sex','Embarked']]
y = df.Survived
```

```python
columns_trans = make_column_transformer(
    (OneHotEncoder(),['Sex','Embarked']),
    remainder='passthrough')

logreg = LogisticRegression(solver='lbfgs')
```

```python
pipe = make_pipeline(columns_trans, logreg )
```

```python
cross_val_score(pipe, X, y, cv=5, scoring='accuracy').mean()
```

```python
ohe = OneHotEncoder(sparse=False)
results = ohe.fit_transform(df[['Sex']])

df = df.join(pd.DataFrame(results, columns = ohe.categories_))
```

```python
results = ohe.fit_transform(df[['Embarked']])

df = df.join(pd.DataFrame(results, columns = ohe.categories_))
```
