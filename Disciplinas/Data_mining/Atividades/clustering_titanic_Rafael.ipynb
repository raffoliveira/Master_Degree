```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from collections import Counter
from sklearn.cluster import KMeans,DBSCAN
from impyute.imputation.cs import mice
from yellowbrick.cluster import KElbowVisualizer
from sklearn.metrics import silhouette_score, homogeneity_completeness_v_measure
```

## Carregando os dados

```python
train = pd.read_csv('../Datasets/train_titanic.csv')
```

```python
train
```

## RemoÃ§Ã£o de atributos

RemoÃ§Ã£o dos atributos **Name, Ticket** e **Cabin** por terem alta cardinalidade.
ApÃ³s a remoÃ§Ã£o o conjunto Ã© dividido:
+ Atributos -> **X**
+ Alvo (***Survived***) -> **y**

```python
#removendo o atributo Survived
df = train.drop(['Name','Ticket','Cabin'], axis=1)
X, y = df.drop(['Survived'],axis=1), df[['Survived']]
```

## Tratamento de atributos categÃ³ricos com ***one hot encoding***

Utilizando dummies do pandas para codificar os atributos **Sex** e ** Embarked**

```python
df=pd.get_dummies(X)
```

```python
df
```

## Tratamento de valores ausentes

Utilizarei a tÃ©cnica de imputaÃ§Ã£o **Multiple imputation by chained equations (MICE)** para o atributo **Age**. Para os atributos **Embarked** e **Fare** serÃ¡ utilizado a remoÃ§Ã£o dos valores ausentes, uma vez que a presenÃ§a Ã© bem pequena.

A tÃ©cnica MICE, em vez de utilizar apenas uma imputaÃ§Ã£o, preenche os valores ausentes utilizando vÃ¡rias imputaÃ§Ãµes de forma iterativa atÃ© uma conversÃ£o. 

Mais informaÃ§Ãµes sobre essa tÃ©cnica pode ser encontrada neste [artigo](https://www.jstatsoft.org/article/view/v045i03/v45i03.pdf).

Abaixo segue uma animaÃ§Ã£o do funcionamento da tÃ©cnica para trÃªs variÃ¡veis. ReferÃªncia: [medium](https://medium.com/@ofirdi/mice-is-nice-but-why-should-you-care-e66698f245a3)

<img src="mice.gif" width=600 height=600 /> 

```python
#verificando valores nulos
X.isnull().sum()
```

```python
#aplicando o MICE 
imputed_values = mice(df.values)

#convertendo o resultado em dataframe
df_ = pd.DataFrame(imputed_values, columns=df.columns)
```

```python
#verificando novamente a presenÃ§a de valores nulos
df_.isnull().sum()
```

```python
#dataframe apÃ³s o one hot encoding
df_
```

## NormalizaÃ§Ã£o dos dados

Utilizarei o Zcore de acordo com a fÃ³rmula abaixo:

$X' = \dfrac{X - \mu}{\sigma}$

onde $X$ Ã© o valor original, $\mu$ Ã© a mÃ©dia dos valores e $\sigma$ Ã© o desvio padrÃ£o dos valores.

```python
#aplicando o zscore para normalizaÃ§Ã£o
df_scaled = stats.zscore(df_)
```

```python
df_scaled
```

## MÃ©todo Elbow

Encontrando o melhor $k$ (nÃºmero de clusters) de acordo com o conjunto de dados utilizando como mÃ©trica de avaliaÃ§Ã£o o **silhouette**. A avaliaÃ§Ã£o Ã© realizada no intervalo de [2,15].
Como observamos abaixo, o melhor $k$ Ã© 2.

```python
model = KMeans()
visualizer_s = KElbowVisualizer(model, k=(2,15), metric='silhouette')
visualizer_s.fit(df_)
visualizer_s.show()
```

# Aplicando a 1Âª tÃ©cnica: ***KMeans*** 

Essa tÃ©cnica basea-se na divisÃ£o do conjunto de dados em $k$ grupos (clusters) onde cada amostra pertence ao grupo em relaÃ§Ã£o a mÃ©dia das amostras  pertencente ao grupo (centrÃ³ides). O ponto principal Ã© a escolha do melhor valor de $k$. Para isso, foi utilizado o mÃ©todo ''joelho'' (**elbow**) para a escolha do melhor valor de $k$ indicando $k=2$

```python
#configurando o numero de clusters e aplicando no conjunto de dados
kmeans = KMeans(n_clusters=2,init='k-means++',random_state=1, n_init=50).fit(df_scaled)

#plotando o resultado
plt.figure(figsize=(9,6))
plt.scatter(df_scaled['Age'], df_scaled['Fare'], c=kmeans.labels_, cmap='plasma')
plt.scatter(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,5], s=100, c='red')
```

# Aplicando a 2Âª tÃ©cnica: ***DBSCAN***

EstÃ¡ tÃ©cnica basea-se no agrupamento por densidade espacial, expandindo o agrupamento pelas amostras de maior densidade

```python
dbscan = DBSCAN().fit(df_scaled)

#plotando o resultado
plt.figure(figsize=(9,6))
plt.scatter(df_scaled['Age'], df_scaled['Fare'], c=dbscan.labels_, cmap='plasma')
#plt.scatter(dbscan.cluster_centers_[:,2], dbscan.cluster_centers_[:,5], s=100, c='red')
```

# Avaliando os resultados

## Silhouette coefficient

O silhouete coeficiente varia de -1 a 1. A funÃ§Ã£o **silhouette_score** retorna o valor mÃ©dio do coeficiente para todas as amostras.

```python
silhouette_k = silhouette_score(df_scaled, kmeans.labels_, random_state=1)
silhouette_d = silhouette_score(df_scaled, dbscan.labels_, random_state=1)
    
print(f'Coeficiente de silhouette utilizando KMeans: {silhouette_k:.4f}')
print(f'Coeficiente de silhouette utilizando DBSCAN: {silhouette_d:.4f}')
```

Logo, para o conjunto de dados do Titanic utilizando o KMeans, o coeficiente de silhouette ficou em 0.26, positivo, indicando um bom agrupamento de forma geral. Enquanto que para o DBSCAN, o coeficiente de silhouette ficou em -0.18, indicando um mau agrupamento dos dados ou criaÃ§Ã£o de mais de dois clusters.

## Homogeneity, Completeness e V_measure

+ ***Homogeneity***: indica se os grupos/clusters contÃ©m membros pertencentes a uma mesma classe
+ ***Completeness***: indica se membros de uma mesma classe estÃ£o no mesmo grupo/cluster.
+ ***V_measure***: Ã© a mÃ©dia harmÃ´nica entre homogeneity e completeness

Para ambas as mÃ©trica, a variaÃ§Ã£o Ã© de 0 a 1, e quanto maior o valor, melhor o agrupamento.

```python
homo_k, compl_k, v_meas_k = homogeneity_completeness_v_measure(y.values.squeeze(),kmeans.labels_)
homo_d, compl_d, v_meas_d = homogeneity_completeness_v_measure(y.values.squeeze(),dbscan.labels_)
   
print(f'MÃ©tricas utilizando KMeans:')
print(f'+ Homogeneity: {homo_k:.4f}')
print(f'+ Completeness: {compl_k:.4f}')
print(f'+ V_measure: {v_meas_k:.4f}')
print('\n')
print(f'MÃ©tricas utilizando DBSCAN:')
print(f'+ Homogeneity: {homo_d:.4f}')
print(f'+ Completeness: {compl_d:.4f}')
print(f'+ V_measure: {v_meas_d:.4f}')
```

Os resultados mostram que as mÃ©tricas para o KMeans nÃ£o foram tÃ£o altas, ambas tendo valores de 0.23. JÃ¡ para o DBSCAN, as mÃ©tricas tambÃ©m ficaram baixas, principalmente a completeness, indicando que o agrupamento nÃ£o estÃ¡ separando bem membros de uma mesma classe, o que consequentemente influencia no valor de V-measure. Logo, juntamente com as mÃ©tricas analisadas, o KMeans apresentou o melhor desempenho de agrupamento em relaÃ§Ã£o ao DBSCAN.

Realizando a anÃ¡lise das mÃ©tricas para o DBSCAN Ã© possÃ­vel observar que as mesmas sÃ£o influenciadas pelo parÃ¢metro $eps$, o qual controla a distÃ¢ncia mÃ¡xima entre amostras para serem consideradas vizinhas uma da outra. Assim, outra anÃ¡lise serÃ¡ levada em conta variando esse parÃ¢metro no intervalo de 0.1 a 5 e retirando a mÃ©dia.

```python
dbscan_silhouette = []
homo_d = []
compl_d = []
v_meas_d = []

values = np.arange(0.1,5,0.1)
#calculando a variaÃ§Ã£o para o DBSCAN
for i in values:
    dbscan_ = DBSCAN(eps=i).fit(df_scaled)
    silhouette_d = silhouette_score(df_scaled, dbscan_.labels_, random_state=1)
    dbscan_silhouette.append(silhouette_d)
    
    homo, compl, v_meas = homogeneity_completeness_v_measure(y.values.squeeze(),dbscan_.labels_)
    homo_d.append(homo)
    compl_d.append(compl)
    v_meas_d.append(v_meas)

print('MÃ©tricas de avaliaÃ§Ã£o do DBSCAN variando eps')
print(f'Coeficiente de silhouette: {np.array(dbscan_silhouette).mean():.4f}')
print(f'Homogeneity: {np.array(homo_d).mean():.4f}')
print(f'Completeness: {np.array(compl_d).mean():.4f}')
print(f'V-measure: {np.array(v_meas_d).mean():.4f}')
```

Observa-se que o coefficiente de silhouette melhorou, indo de -0,18 para 0,25. JÃ¡ as outras mÃ©tricas, nÃ£o houve melhoras de resultado.
Logo a variaÃ§Ã£o do $eps$ interferiu apenas no coeficiente de silhouette.
